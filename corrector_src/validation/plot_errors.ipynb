```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import sys
import os
```

```python
df_non_enhanced = pd.read_csv("error_calculation.csv")
df_enhanced = pd.read_csv("ml_enhanced_loss.csv")
df_non_enhanced["origin"] = ["NE" * len(df_non_enhanced)]
df_enhanced["origin"] = ["E" * len(df_enhanced)]
df = pd.concat([df_non_enhanced, df_enhanced])
```

```python
x = "rand_1"
plt.figure(figsize=(8, 6))
sns.lineplot(data=df, x=x, y="loss", hue = 'origin', marker="o")

# Ensure x-axis only shows integer ticks
x_ticks = sorted(df[x].unique())
plt.xticks(x_ticks)
plt.yscale("log")
plt.xlabel(x)
plt.ylabel("Loss")
#plt.title("Validation Loss vs Residual Blocks (Grouped by Modes)")
plt.grid(False)
plt.tight_layout()
plt.show()
```
